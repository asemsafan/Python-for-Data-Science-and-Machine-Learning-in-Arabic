# Distribution Plots

```python
# importing seaborn
import seaborn as sns
%matplotlib inline
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

### 1- distplot #deprecated

```python
# sns.distplot (tips_df['total_bill'],kde=False) #warning 
```

## 1- Displot

https://seaborn.pydata.org/generated/seaborn.displot.html

```python
sns.displot (data=tips_df, x='total_bill')
```

```python
sns.displot (data=tips_df, x='total_bill',kind='kde')
```

```python
sns.displot (data=tips_df, x='total_bill',kde=True)
```

## 2- Joint plot

```python
sns.jointplot (x= 'total_bill' , y='tip' ,data=tips_df )
```

```python
sns.jointplot (x= 'total_bill' , y='tip' ,data=tips_df,kind='scatter' )
```

```python
sns.jointplot (x= 'total_bill' , y='tip' ,data=tips_df,kind='reg' )
```

```python
sns.jointplot (x= 'total_bill' , y='tip' ,data=tips_df,kind='hex' )
```

## 3- Pair plot

https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot

```python
sns.pairplot (tips_df)
```

```python
sns.pairplot (tips_df,hue='sex')
```

## 4- Rug plot

```python
sns.rugplot (tips_df['total_bill'])
```

## 5- Kde plot

```python
sns.kdeplot (tips_df['total_bill'])
```
