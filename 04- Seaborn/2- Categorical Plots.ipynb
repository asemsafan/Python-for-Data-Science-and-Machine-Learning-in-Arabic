# Categorical Plots

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
#https://github.com/mwaskom/seaborn-data
tips.head()
```

### 1- Barplot `sns.barplot()`

```python
sns.barplot(x='sex', y='total_bill',data=tips)
```

```python
import numpy as np
```

```python
sns.barplot(x='day', y='total_bill',data=tips,estimator=np.std,hue='sex')
```

### 2- Countplot `sns.countplot()`

```python
sns.countplot (x='sex',data=tips)
```

```python
sns.countplot (x='day',data=tips)
```

```python
sns.countplot (x='day',data=tips,hue='sex')
```

### 3- Boxplot `sns.boxplot()`

```python
sns.boxplot (x='day', y='total_bill',data=tips)
```

```python
sns.boxplot (x='day', y='total_bill',data=tips,hue='sex')
```

```python
sns.boxplot (x='day', y='total_bill',data=tips,hue='sex',palette="rocket")
```

### 4- Violinplot `sns.violinplot()`

```python
sns.violinplot (x='day', y='total_bill',data=tips)
```

```python
sns.violinplot (x='day', y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot (x='day', y='total_bill',data=tips,hue='sex',split=True)
```

### 5- Stripplot `sns.stripplot()`

```python
sns.stripplot (x='day', y='total_bill',data=tips)
```

```python
sns.stripplot (x='day', y='total_bill',data=tips,hue='smoker',jitter=True)
```

```python
sns.stripplot (x='day', y='total_bill',data=tips,hue='smoker')
```

```python
# sns.stripplot (x='day', y='total_bill',data=tips,hue='smoker',split=True)
```

### 6- Swarmplot `sns.swarmplot()`

```python
sns.swarmplot (x='day', y='total_bill',data=tips)
```

```python
sns.swarmplot (x='day', y='total_bill',data=tips,color='white')
sns.violinplot (x='day', y='total_bill',data=tips)
```
