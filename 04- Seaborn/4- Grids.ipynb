# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# https://github.com/mwaskom/seaborn-data
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].unique()
```

```python
penguins['island'].unique()
```

```python
sns.pairplot(penguins)
```

```python
sns.pairplot(penguins,hue='species',palette='tab10')
```

### 1- seaborn.PairGrid `sns.PairGrid()`

https://seaborn.pydata.org/generated/seaborn.PairGrid.html

```python
sns.PairGrid(penguins)
```

```python
g = sns.PairGrid(penguins)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(penguins)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

### 2- seaborn.Facet Grid `sns.FacetGrid()`

https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python
penguins.head()
```

```python
penguins['species'].unique()
```

```python
sns.FacetGrid(penguins, row='species' ,col='sex' )
```

```python
g = sns.FacetGrid(penguins, row='sex' ,col='species' )
g.map(plt.hist,'body_mass_g')
```

```python
g = sns.FacetGrid(penguins, row='sex' ,col='species' )
g.map(plt.scatter,'body_mass_g','flipper_length_mm')
```
