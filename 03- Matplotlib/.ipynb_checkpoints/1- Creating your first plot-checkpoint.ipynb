#### 3- your first plot

```python
plt.plot (ages_x,dev_y)
plt.title ('Average Salary by Age')
plt.xlabel ('Age by Year')
plt.ylabel ('Average Salary ($)')
plt.show()
```

#### 4- Two lines in the same plot with legends, title, xlabel and ylabel

```python
plt.plot (ages_x,dev_y)
plt.plot (ages_x,py_dev_y)
# adding legned 
plt.legend (['All Developers','Python Devs'])
plt.show ()
```

```python
# adding legend by lebeling every plot
plt.plot (ages_x,dev_y,label ='All Developers')
plt.plot (ages_x,py_dev_y, label ='Python Devs')
# adding graph title
plt.title ('Average Salary by Age')
# adding x axis label
plt.xlabel ('Age by Year')
# adding y axis label
plt.ylabel ('Average Salary ($)')
plt.legend ()
plt.show ()
```

```python
# Editing lines formatting 
# the link used in the video
# https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot (ages_x,dev_y,color= '#0dd69d',marker ='o',linestyle='--', label ='All Developers')
plt.plot (ages_x,py_dev_y, '^:y' , label ='Python Devs')

plt.legend ()
plt.show ()
```

#### 5- selecting different plt style

```python
print (plt.style.available)
```

```python
plt.style.use('fivethirtyeight')
plt.plot (ages_x,dev_y,label ='All Developers')
plt.plot (ages_x,py_dev_y, label ='Python Devs')
plt.title ('Average Salary by Age')
plt.xlabel ('Age by Year')
plt.ylabel ('Average Salary ($)')
plt.legend()
plt.show()
```

#### 6- Using `plt.xkcd()`

```python
plt.xkcd ()
plt.plot (ages_x,dev_y,label ='All Developers')
plt.plot (ages_x,py_dev_y, label ='Python Devs')
plt.title ('Average Salary by Age')
plt.xlabel ('Age by Year')
plt.ylabel ('Average Salary ($)')
plt.legend()
plt.show()
```

#### 7- removing xkcd style by `plt.rcdefaults()`

```python
# the first line remove the effect of plt.xkcd ()
plt.rcdefaults()
plt.plot (ages_x,dev_y,label ='All Developers')
plt.show()
```
