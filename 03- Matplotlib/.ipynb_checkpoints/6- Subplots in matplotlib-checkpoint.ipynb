# Subplots in matplotlib

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('subplots_data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

```python
plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')
plt.legend()

plt.title('Median Salary by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary')
plt.show()
```

### Introductin to OOB method

```python
fig ,axes = plt.subplots()

axes.plot(ages, py_salaries, label='Python')
axes.plot(ages, js_salaries, label='JavaScript')
axes.legend()

axes.set_title('Median Salary by Age')
axes.set_xlabel('Ages')
axes.set_ylabel('Median Salary')
plt.show()
```

```python
fig , (ax1,ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')
ax1.legend()

ax1.set_title('Median Salary by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary')

ax2.set_title('Median Salary by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary')

plt.tight_layout()

plt.show()
```

```python
fig , axes = plt.subplots(nrows=2,ncols=1,sharex=True)

axes[0].plot(ages, py_salaries, label='Python')
axes[1].plot(ages, js_salaries, label='JavaScript')
axes[0].legend()

axes[0].set_title('Median Salary by Age')
axes[0].set_xlabel('Ages')
axes[0].set_ylabel('Median Salary')

axes[1].set_title('Median Salary by Age')
axes[1].set_xlabel('Ages')
axes[1].set_ylabel('Median Salary')

plt.tight_layout()

plt.show()
```
