# Scatter Plot

https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
np.random.seed(42)
```

```python
x_values = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y_values = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
y2= [i + np.random.randint (2,8) for i in x_values]
y3= [i * (- np.random.randint (2,8)) for i in x_values]

print (y2)
print (y3)
```

```python
plt.scatter (x_values, y_values)
plt.show()
```

```python
plt.scatter (x_values, y2)
plt.show()
```

```python
plt.scatter (x_values, y3)
plt.show()
```

```python
plt.scatter (x_values, y2,s=100,c='red',edgecolors='black',linewidths=2)
# marker='X',alpha=.5 > as argument 
plt.show()
```

```python
colors = [7, 5, 9, 7, 5, 7, 2, 5, 3, 7, 1, 2, 8, 1, 9, 2, 5, 6, 7, 5]

sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174,
         538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter (x_values, y2,s=sizes,c=colors,edgecolors='black',linewidths=1)
plt.show()
```

```python
# link for cmap => https://matplotlib.org/3.5.0/tutorials/colors/colormaps.html
plt.scatter (x_values, y2,s=sizes,c= colors ,cmap='RdPu' ,edgecolors='black',linewidths=1)
cbar = plt.colorbar ()
cbar.set_label ('Reviews')
plt.show()
```

```python
df = pd.read_csv('scatter_data.csv')
```

```python
df.head()
```

```python
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

```python
plt.scatter (likes,view_count)
plt.show()
```

```python
plt.scatter (likes,view_count,c=ratio,cmap='YlGn')
cbar = plt.colorbar ()
cbar.set_label ('Ratio')
plt.xscale ('log')
plt.yscale ('log')
plt.title('Trending Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.show()
```
